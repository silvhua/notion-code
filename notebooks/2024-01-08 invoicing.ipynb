## 1.3

```python
from data_viz import *
    
def get_invoice_records(df, start_date, end_date, filter_dict):
    classified_df = classify_projects(df)

    period_df = filter_by_period(
        classified_df, column='created_time', period=None, start_date=start_date, end_date=end_date
    )
    client_df = filter_df_all_conditions(period_df, filter_dict, verbose=False)
    return client_df

def time_per_task(client_df, task_column='Task Name', unbilled_column='Unbilled'):
    if client_df[unbilled_column].dtype == 'bool':
        client_df[unbilled_column] = client_df[unbilled_column].apply(
            lambda x: 'Billed' if x else 'Unbilled'
        )
    df = pd.DataFrame(client_df.groupby([task_column, unbilled_column])['Elapsed'].agg('sum'))

    df = df.unstack()
    df.columns = df.columns.droplevel()
    df.columns.name = None
    df = df.reset_index()
    df = df.fillna(0)
    return df

filter_dict = {
    'Task Project name': ['Coach McLoone', 'GHL Chatbot']
}

start_date = '2024-01-02'
end_date = '2024-01-03'

# get_invoice_records(df, start_date, end_date, filter_dict)
time_per_task(get_invoice_records(df, start_date, end_date, filter_dict))
```
