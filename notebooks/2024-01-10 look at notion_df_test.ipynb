```python
def invoice_timesheet(df, include_roadmap_item=True):
    df['Date'] = df['Name'].str.extract(r'(\d{4}-\d{2}-\d{2})', expand=False)
    invoice_columns = [
        'Date',
        'Task Project name',
        'start time',
        'end time',
        'Elapsed',
        'Task Name',
        'Notes'
    ]
    if include_roadmap_item:
        invoice_columns.insert(2, 'Roadmap Item')
    return df[invoice_columns]

invoice_timesheet(df)
```

```python
def Invoice_Timesheet(df):
    df['Date'] = df['Name'].str.extract(r'(\d{4}-\d{2}-\d{2})', expand=False)
    invoice_columns = [
        'Date',
        'Task Project name',
        'start time',
        'end time',
        'Elapsed',
        'Task Name'
    ]
    # if include_roadmap_item:
    #     invoice_columns.insert(2, 'Roadmap Item')
    if df['Roadmap Item'].apply(lambda x: len(x)>0).sum() > 0:
        invoice_columns.insert(2, 'Roadmap Item')
    if 'Notes' in df.columns:
        invoice_columns.append('Notes')
    df = df[invoice_columns].round(2)
    df.columns = [f'| {column}' for column in df.columns]
    return df


Invoice_Timesheet(df)
```
